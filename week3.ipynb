**lets see correlation between different between dataset using heatmap**

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', annot_kws={"size": 10})
plt.show()
```

from above heatmap some deductions are
1. gamma ray is postively corelated with density and poisson ratio.poisson ratio is ratio of stress/strain of material.
2. resistivity of material will increase linearly with density because both have postive corealtion.
3. density have postive correlation with vp and vs
4. density porossity and possion ratio both have positive correlation with Delta T
5. classification have positive corelation with vs, density calculated and density
