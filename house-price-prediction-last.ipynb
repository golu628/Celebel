```python
# capturing hidden numeric and hidden categorical columns 

def grab_col_names(dataframe, cat_th=10, car_th=20):
    """
    grab_col_names for given dataframe

    :param dataframe:
    :param cat_th:
    :param car_th:
    :return:
    """

    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]

    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]

    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]

    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')

    # cat_cols + num_cols + cat_but_car = number of variables.
    # num_but_cat is already inside cat_cols.
    # so all variables will be selected with all these 3 lists: cat_cols + num_cols + cat_but_car
    # num_but_cat is provided for reporting only.

    return cat_cols, cat_but_car, num_cols

# Applying the df dataset through the grab_col_names function 
cat_cols, cat_but_car, num_cols = grab_col_names(df)
```

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.70):
    # Let's select only numeric columns
    numeric_cols = dataframe.select_dtypes(include=['number'])
    corr = numeric_cols.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(np.bool_))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]

    if plot:
        plt.figure(figsize=(15, 10))
        sns.heatmap(corr, cmap="coolwarm", annot=True, linewidths=.5)
        plt.show()
    
    return drop_list

# Find highly correlated columns
high_corr_cols = high_correlated_cols(df, plot=True)

print("High Correlation Columns:", high_corr_cols)

# Variables with high correlation
# ['1stFlrSF', 'TotRmsAbvGrd', 'GarageYrBlt', 'GarageArea', 'SalePrice']
```

```python
# Suppression of outliers
def outlier_thresholds(dataframe, variable, low_quantile=0.10, up_quantile=0.90):
    quantile_one = dataframe[variable].quantile(low_quantile)
    quantile_three = dataframe[variable].quantile(up_quantile)
    interquantile_range = quantile_three - quantile_one
    up_limit = quantile_three + 1.5 * interquantile_range
    low_limit = quantile_one - 1.5 * interquantile_range
    
    return low_limit, up_limit


# suppression is the fixation of outliers with respect to the minimum and maximum values

# We can set our threshold according to our data set, according to the number of observations, according to the structure of the variables, according to our own know-how
# generally taken as 75% by 25%. but it is possible to increase these values in order not to lose too much information
# too much suppression can often lead to loss of information and loss of correlation between variables
```

```python
# Outlier check

# According to these thresholds, we will check whether there are outliers in the variables, and if so, which ones
# we will query the outliers of a variable as bool
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False


for col in num_cols:
    if col != "SalePrice":
      print(col, check_outlier(df, col))
```

```python
# Suppression of outliers
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit


for col in num_cols:
    if col != "SalePrice":
        replace_with_thresholds(df,col)
```

```python
# with this function, when we receive a data, we can quickly analyze this data with what is the frequency of deficiencies
# we will see which variables are missing and what is the rate of these missing variables

def missing_values_table(dataframe, na_name=False):
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]

    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)

    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)

    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])

    print(missing_df, end="\n")

    if na_name:
        return na_columns

missing_values_table(df)
```

How to solve the missing data problem

###########################################

**Deletion : deleting variables of missing data**

We can delete it by saying dropna, but this will reduce the number of observations.
missing values can be deleted if the number of missing observations is small, but missing values can be deleted if the number of observations is small or the missing values
deleting data where there is a lot of it will lead to a significant loss of data



**Value assignment methods: simple assignment methods such as mean mode median**

We can fill in the gaps in the variables with the median or the mean, or we can fill them with any constant value.


**Predictive methods: Assigning values according to predictions by ML or some statistical methods**


################################################################################

we will now switch to assignment with mode median methods

completion of missing numeric variables


The default value of this function is the median. we can change this later by entering num_method="XXX", which allows missing values to be filled with median or mean

For categorical variables, the threshold is set to 20, which refers to the maximum number of unique classes this variable can have. 
The default target value is "SalePrice".

The function first identifies variables with missing values in the dataset and stores them in a list called variables_with_na. It then stores the target variable in a temporary variable called temp_target.

The function then uses the Apply() method to apply a lambda function to each column in the dataset.

The lambda function checks the data type and the number of unique values of each column and fills in the missing values as follows:

If the data type is "O" (i.e. object) and the number of unique values is less than or equal to cat_length=20, missing values are assigned with mod (i.e. the most frequent value).

If num_method is "mean", missing values in columns with data type other than "O" are associated with the mean value.

If num_method is "median", missing values in columns with data type other than "O" are associated with the median value.

Finally, the function restores the target variable and prints the number of missing values in each column before and after the assignment.

It then returns the modified data set.

```python
# This function allows missing values to be filled with median or mean

def quick_missing_imp(data, num_method="median", cat_length=20, target="SalePrice"):
    variables_with_na = [col for col in data.columns if data[col].isnull().sum() > 0]  # Variables with missing values are listed

    temp_target = data[target]

    print("# BEFORE")
    print(data[variables_with_na].isnull().sum(), "\n\n")  # Number of missing values of variables before implementation

    # fill empty values with mode if the number of variable objects and classes is less than or equal to cat_lengthe
    data = data.apply(lambda x: x.fillna(x.mode()[0]) if (x.dtype == "O" and len(x.unique()) <= cat_length) else x, axis=0)

    # num_method mean fills empty values of variables whose type is not object with mean
    if num_method == "mean":
        data = data.apply(lambda x: x.fillna(x.mean()) if x.dtype != "O" else x, axis=0)
    # If num_method is median, empty values of variables whose type is not object are filled with the mean
    elif num_method == "median":
        data = data.apply(lambda x: x.fillna(x.median()) if x.dtype != "O" else x, axis=0)

    data[target] = temp_target

    print("# AFTER \n Imputation method is 'MODE' for categorical variables!")
    print(" Imputation method is '" + num_method.upper() + "' for numeric variables! \n")
    print(data[variables_with_na].isnull().sum(), "\n\n")

    return data


df = quick_missing_imp(df, num_method="median", cat_length=17)
```

# Let's analyze rare and apply rare encoder

```python
# we need to select the categorical variables here we will call grab col names
# retrieving categorical variables
# why are we doing this
# let's not have an unnecessary number of categories and let's put similar categories together or remove the useless ones
cat_cols, num_cols, cat_but_car = grab_col_names(df)
```

```python
# Now let's analyze these categorical variables and their classes according to the rarity of the classes

# Analyzing the scarcity and multiplicity of categorical variables.
######################################################################
# Now here we need a function and we need to do something with this function in such a way that
# get the classes of categorical variables and the proportions of these classes, if we say plot=true we also get the graphs


def cat_summary(dataframe, col_name, plot=False):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    print("##########################################")
    if plot:
        sns.countplot(x=dataframe[col_name], data=dataframe)
        plt.show()


for col in cat_cols:
    cat_summary(df, col)

# Why did we do it

# rare means rarely observed rare
# little observed variables will not contain much information in the column when we do one hot encoding
# therefore they will not find a correspondence in the modeling
# therefore the values that we pass through one hot encoder and turn into variables should also have measurement quality
# and we want to consider various possible effects on the dependent variable.
# so we can use the rare encoder to get rid of unnecessary variables
# To summarize, to get rid of low values in the classes of a categorical variable in a dataset
# we set a threshold value, and the classes that fall below this particular threshold value
# we put them together and we call them rare, so in a way we leave them out
```

```python
# Analyzing the relationship between rare categories and the dependent variable.

# What are the effects of the variables of the classes we will now include in this rare category on the dependent variable and
# We will analyze what is the relationship between

# Why are we doing this? Let's not have an unnecessary number of categories, let's group together as many similar places, variables or categories as possible
# Let's get rid of uninformative classes or put them in another category

# with this function the frequencies of the classes will be averaged in terms of ratios and target, i.e. SalePrice

# Analyzing the distribution of categorical columns

def rare_analyser(dataframe, target, cat_cols):
    for col in cat_cols:
        print(col, ":", len(dataframe[col].value_counts()))
        print(pd.DataFrame({"COUNT": dataframe[col].value_counts(),
                            "RATIO": dataframe[col].value_counts() / len(dataframe),
                            "TARGET_MEAN": dataframe.groupby(col)[target].mean()}), end="\n\n\n")

rare_analyser(df, "SalePrice", cat_cols)
```

```python
# Writing the rare encoder.

# we will set our rare percentage and it will aggregate the classes of categorical variables below this percentage.
# our rare encoder will collect and aggregate the rare classes of categorical variables in the dataset.
# gives them the designation rare

# Identification of rare classes
def rare_encoder(dataframe, rare_perc):
    temp_df = dataframe.copy()

    rare_columns = [col for col in temp_df.columns if temp_df[col].dtypes == 'O'
                    and (temp_df[col].value_counts() / len(temp_df) < rare_perc).any(axis=None)]

    for var in rare_columns:
        tmp = temp_df[var].value_counts() / len(temp_df)
        rare_labels = tmp[tmp < rare_perc].index
        temp_df[var] = np.where(temp_df[var].isin(rare_labels), 'Rare', temp_df[var])

    return temp_df.head()


rare_encoder(df, 0.01)
```

**Standard Deviation**
The standard deviation is a measure of how much the data points differ from the mean. That is, it shows how widespread the distribution of values in the data set is. The larger the standard deviation, the more the data points deviate from the mean and the wider the distribution.

**Difference and Comparison between Mean and Standard Deviation**
The mean represents the central tendency of the dataset, while the standard deviation represents the average distance of the data points from this center. Together, they help us understand the overall structure and distribution of the data set.

**Low Standard Deviation:** If the standard deviation value is low, this indicates that the data points are close to the mean and the data set is fairly homogeneous. That is, the data points are similar and consistent.

**High Standard Deviation:** A high standard deviation indicates that the data points deviate greatly from the mean and the data set is heterogeneous. There may be large differences between data points and the data set is more variable.

**Comparison Method:** A direct "difference" calculation is usually not done to compare the mean and standard deviation. Instead, we evaluate the magnitude of the standard deviation relative to the mean. For example, whether the standard deviation is large or small relative to the mean gives information about the spread of the data distribution.

Log transformation and its inverse log transformation is a common technique in data science, especially in regression models. The main reasons for using these techniques are as follows:

**Correcting Data Distribution:** Many data sets encountered in the real world show deviations from the normal distribution. In particular, cases where the target variable is skewed to the right or left can violate the assumptions of some algorithms such as linear regression. Log transformation can help these algorithms perform better by transforming skewed data into a more normal distribution.

**Fixing the Variance of Error Terms:** Another important assumption for regression models is that the error terms have a constant variance (homoscedasticity). In skewed data, observations with large values can often have larger error terms. Log transformation fixes this variance, allowing the model to make more consistent predictions.

**Dealing with Very Large or Very Small Values:** In some cases, the target variable may have very large or very small values. Such values can negatively affect the model's learning process. Log transformation can alleviate this problem by compressing the range of values.

*Once the data is log-transformed, the model's predictions are also made on the log-transformed target variable. However, in real world applications, the forecasts need to be converted to the original scale. Therefore, the predictions at the output of the model need to be converted to the original scale by taking the inverse of the log transformation.* This is accomplished with the function np.expm1. The function np.expm1(x) calculates exp(x) - 1, which is the inverse of the log1p transformation (i.e. log(1+x)).*

*When evaluating the performance of the model, converting predictions and actual values to the original scale makes the resulting error metric (e.g. RMSE) more meaningful and interpretable. This is because end-users or decision makers will want to understand the outputs and performance of the model on the original scale.*

Log dönüşümü ve ardından yapılan inverse log dönüşümü (log dönüşümünün tersi), veri biliminde özellikle regresyon modellerinde sıkça karşılaşılan bir tekniktir. Bu tekniklerin kullanılmasının başlıca nedenleri şunlardır:

**Veri Dağılımını Düzeltmek:** Gerçek dünyada karşılaşılan birçok veri seti, normal dağılımdan sapmalar gösterir. Özellikle, hedef değişkenin sağa ya da sola çarpık olduğu durumlar, lineer regresyon gibi bazı algoritmaların varsayımlarını ihlal edebilir. Log dönüşümü, çarpık veriyi daha normal bir dağılıma dönüştürerek bu algoritmaların daha iyi performans göstermesine yardımcı olabilir.

**Hata Terimlerinin Varyansını Sabitlemek:** Regresyon modelleri için bir diğer önemli varsayım, hata terimlerinin sabit bir varyansa (homoscedasticity) sahip olmasıdır. Çarpık verilerde, büyük değerlere sahip gözlemler genellikle daha büyük hata terimlerine sahip olabilir. Log dönüşümü, bu varyansı sabitleyerek modelin daha tutarlı tahminler yapmasına olanak tanır.

**Çok Büyük veya Çok Küçük Değerlerle Başa Çıkmak:** Bazı durumlarda, hedef değişkende çok büyük veya çok küçük değerler olabilir. Bu tür değerler, modelin öğrenme sürecini olumsuz etkileyebilir. Log dönüşümü, değer aralığını sıkıştırarak bu sorunu hafifletebilir.

*Veriye log dönüşümü uygulandıktan sonra, modelin tahminleri de log dönüşümlü hedef değişken üzerinde yapılmış olur. Ancak, gerçek dünya uygulamalarında tahminlerin orijinal ölçeğe dönüştürülmesi gerekir. Bu nedenle, modelin çıkışındaki tahminlerin log dönüşümünün tersi alınarak orijinal ölçeğe dönüştürülmesi gerekir.* Bu işlem, np.expm1 fonksiyonu ile gerçekleştirilir. np.expm1(x) fonksiyonu, exp(x) - 1 hesaplamasını yapar ve bu, log1p dönüşümünün (yani log(1+x)) tersidir.

*Modelin performansını değerlendirirken, tahminlerin ve gerçek değerlerin orijinal ölçeğe dönüştürülmesi, elde edilen hata metriğinin (örneğin, RMSE) daha anlamlı ve yorumlanabilir olmasını sağlar. Çünkü son kullanıcılar veya karar vericiler, modelin çıktılarını ve performansını orijinal ölçekte anlamak isteyecektir.*

# Ranking of features with feature_importance

```python
# feature importance
def plot_importance(model, features, num=len(X), save=False):

    feature_imp = pd.DataFrame({"Value": model.feature_importances_, "Feature": features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False)[0:50])
    plt.title("Features")
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig("importances.png")
```
